```python
import numpy as np
import matplotlib.pyplot as plt

fs = 50  # Frequenza di campionamento
t = np.arange(0, 1, 1/fs)
signal = (np.cos(2*np.pi*4*t)**2)*np.sin(2*np.pi*8*t) 

signal_fft = np.fft.fft(signal)
N = len(signal)
freq = np.fft.fftfreq(N, 1/fs)
amplitude = np.abs(signal_fft) / N
phase = np.angle(signal_fft)

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(freq[:N//2], amplitude[:N//2])
plt.title('Spettro di Ampiezza')
plt.subplot(2, 1, 2)
plt.plot(freq[:N//2], phase[:N//2])
plt.title('Spettro di Fase')
plt.show()
```
